```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_excel("Data.xlsx")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.replace(to_replace=True, value=1, inplace=True)
df.replace(to_replace=False, value=0, inplace=True)
df.replace(to_replace=1.0, value=1, inplace=True)
df.replace(to_replace=0.0, value=0, inplace=True)
```

```python
df['hour17'].fillna(df['hour17'].median(), inplace=True)
df['hour18'].fillna(df['hour18'].median(), inplace=True)
df['hour19'].fillna(df['hour19'].median(), inplace=True)
df['hour20'].fillna(df['hour20'].median(), inplace=True)
df['hour21'].fillna(df['hour21'].median(), inplace=True)
df['hour22'].fillna(df['hour22'].median(), inplace=True)
df['hour23'].fillna(df['hour23'].median(), inplace=True)
```

```python
columns_to_convert = ['hour17', 'hour18', 'hour19', 'hour20', 'hour21', 'hour22', 'hour23']
for col in columns_to_convert:
    df[col] = df[col].astype(int)

```

```python
selected_columns = ['hour0', 'hour1', 'hour2', 'Station_Type', 'Station_Category']
subset_df = df[selected_columns]
```

```python
daily_aggregated_df = df.groupby('Ping_Date').sum()
```

```python
daily_aggregated_df['Total_Uptime'] = daily_aggregated_df.iloc[:, 6:].sum(axis=1)
daily_aggregated_df['Total_Uptime'].plot(kind='line', marker='o', figsize=(12, 6))
plt.title('Daily Uptime Analysis')
plt.xlabel('Date')
plt.ylabel('Total Uptime')
plt.show()
```

```python
hourly_aggregated_df = df.groupby(df['Ping_Date']).sum().iloc[:, 6:]
hourly_aggregated_df.mean().plot(kind='bar', figsize=(18, 8))
plt.title('Hourly Uptime Analysis')
plt.xlabel('Hour of the Day')
plt.ylabel('Mean Uptime')
plt.show()
```

```python
df['Ping_Date'] = pd.to_datetime(df['Ping_Date'])
```

```python
df['Day_of_Week'] = df['Ping_Date'].dt.dayofweek
```

```python
def label_weekday(row):
    if row['Day_of_Week'] < 5:  # Monday to Friday
        return 'Weekday'
    else:
        return 'Weekend'
```

```python
df['Day_Type'] = df.apply(label_weekday, axis=1)
```

```python
weekday_data = df[df['Day_Type'] == 'Weekday'].iloc[:, 7:-3]
weekend_data = df[df['Day_Type'] == 'Weekend'].iloc[:, 7:-3]
```

```python
weekday_mean = weekday_data.mean()
weekend_mean = weekend_data.mean()
```

```python
plt.figure(figsize=(18, 6))
sns.lineplot(data=pd.DataFrame({'Weekday': weekday_mean, 'Weekend': weekend_mean}))
plt.title('Charger Usage Comparison: Weekdays vs. Weekends')
plt.xlabel('Hour of the Day')
plt.ylabel('Mean Uptime')
plt.show()
```

```python
df['Station_Type']=df['Station_Type'].astype('category').cat.codes
df['Station_Category']=df['Station_Category'].astype('category').cat.codes
df['Station_Access']=df['Station_Access'].astype('category').cat.codes
```

```python
columns_of_interest = ['hour0', 'hour1', 'hour2','hour3','hour4','hour5', 'hour6', 'hour7', 'hour8', 'hour9', 'hour10', 'hour11', 'hour12', 'hour13', 'hour14', 'hour15', 'hour16', 'hour17', 'hour18', 'hour19', 'hour20', 'hour21', 'hour22', 'hour23', 'Station_Type', 'Station_Category', 'Station_Access']
```

```python
subset_df = df[columns_of_interest]
```

```python
from scipy.stats import pointbiserialr
correlation_results = {}
for characteristic in ['Station_Type', 'Station_Category', 'Station_Access']:
    correlation_result = pointbiserialr(subset_df[characteristic], subset_df['hour0'])
    correlation_results[characteristic] = correlation_result.correlation
```

```python
print("Correlation Results:")
for characteristic, correlation in correlation_results.items():
    print(f"{characteristic}: {correlation}")
```

```python
correlation_df = pd.DataFrame(list(correlation_results.items()), columns=['Station_Characteristic', 'Correlation'])
```

```python
subset_df_encoded = pd.get_dummies(subset_df, columns=['Station_Type', 'Station_Category', 'Station_Access'], prefix=['Station_Type', 'Station_Category', 'Station_Access'])
combined_df = pd.concat([subset_df, subset_df_encoded], axis=1)
#subset_df_encoded = pd.get_dummies(subset_df, columns=['Station_Type', 'Station_Category', 'Station_Access'])

# Calculate the correlation matrix
correlation_matrix = subset_df_encoded.corr()

# Create a heatmap
plt.figure(figsize=(28, 26))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```
